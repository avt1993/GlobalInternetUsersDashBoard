```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
giu_df = pd.read_csv('/Users/antoniovillardaga/Documents/Python/GlobalInternetUsersDashBoard/Global Internet Users.csv')
```

```python
giu_df.head()
```

```python
country_name_df = giu_df['Entity'].unique()
```

```python
gui_df_internet_users = giu_df.groupby('Year')['No. of Internet Users'].sum().reset_index()
```

```python
gui_df_internet_users['Increase In Users'] = gui_df_internet_users['No. of Internet Users'].diff()
gui_df_internet_users['Percentage Increase'] = ((gui_df_internet_users['Increase In Users'] / gui_df_internet_users['No. of Internet Users'].shift()) * 100).round()
gui_df_internet_users

gui_df_internet_users
```

```python
gui_df_internet_users.plot(x = 'Year', y = 'No. of Internet Users', kind = 'line')
plt.title('Internet Users from 1980 - 2020')
plt.xlabel('Year')
plt.ylabel('# Internet Users (Billions)')
plt.grid(True)
plt.ylim(0, 9.8e9)
plt.show()
```

```python
usa_df = giu_df[giu_df['Entity'] == "United States"].copy().reset_index()
usa_df
```

```python
mex_df = giu_df[giu_df['Entity'] == "Mexico"].copy().reset_index()
mex_df
```

```python
fig, ax = plt.subplots()

ax.plot(usa_df['Year'], usa_df['No. of Internet Users'], label='USA')

ax.plot(mex_df['Year'], mex_df['No. of Internet Users'], label='Mexico')

plt.title('USA vs Mexico Internet Users')
plt.xlabel('Year')
plt.ylabel('Internet Users')
plt.grid(True)
plt.legend()

plt.show()
```
